```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# read weather history
weather_list = pd.read_csv('weather_history.csv', sep=',')


# months List
months = np.array(['jan' , 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez'])

# maxTemp/month
maxTemp = np.array([29, 29, 28, 27, 23, 23, 23, 25, 25, 26, 27, 28])

# minTemp/month
minTemp = np.array([21, 21, 20, 18, 18, 14, 13, 14, 16, 18, 18, 20])

# rainyDays/month
rainyDays = np.array([14, 12, 11, 6, 6, 4, 3, 3, 7, 8, 9, 12])

# Kilowats/month
kw = np.array([256, 218, 232, 210, 197, 208, 229, 212, 243, 249, 217, 279])
```

```python
l1 = tf.keras.layers.Dense(units=1, input_shape=[1])

model = tf.keras.Sequential([l1])

model.compile(loss='mean_squared_error',optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
history = model.fit(minTemp, kw, epochs=500, verbose=False)
```

```python
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
```

```python
plt.plot(months, minTemp, kw)
```

```python
# Temperatura minma mdia no ms
inputedTemp = 18

# Comando para prever o valor de killowats gastos no respectivo ms
predictedKw = model.predict([inputedTemp])

print(predictedKw)
```
