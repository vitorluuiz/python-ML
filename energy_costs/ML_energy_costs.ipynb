Importing used libs

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Create dataSet training

```python
# read weather history
weather_list = pd.read_csv('weather_history.csv', sep=',')
kw = weather_list['kw'] /10
```

Create model

```python
l1 = tf.keras.layers.Dense(units=1, input_shape=[1])

model = tf.keras.Sequential([l1])

model.compile(loss='mean_squared_error',optimizer=tf.keras.optimizers.Adam(0.001))
```

Training model

```python
history = model.fit(weather_list['maxTemp'], kw, epochs=2000, verbose=False)
```

Show the loss history from training

```python
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
```

Show the relation of minTemp and killowats

```python
# create plot to show the dataSet
plt.plot(
    # add maxTemp in yellow
    weather_list['mês'], weather_list['maxTemp'],'y^',
    # add mintemp in green
    weather_list['mês'],weather_list['minTemp'],'cv',
    # add rainyDays in blue
    weather_list['mês'],weather_list['rainyDays'],'b:',
    # add kw in black
    weather_list['mês'], kw, 'k--')

plt.title('Histórico climático de São Paulo em 2022')

# show plot
plt.show()
```

```python
# Temperatura miníma média no mês
inputedTemp = 28

# Comando para prever o valor de killowats gastos no respectivo mês
predictedKw = model.predict([inputedTemp])

print(predictedKw * 10)
```
